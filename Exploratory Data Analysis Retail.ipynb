### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading Data Set

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

### Shape of Data

```python
df.shape
```

### Data Types

```python
data_types = pd.DataFrame(df.dtypes, columns = ['Types'])
data_types
```

### Missing Values

```python
missing_values = pd.DataFrame(df.isna().sum(), columns = ['Missing Values'])
missing_values
```

### Number of Unique Values in Each Column

```python
for i in df.columns:
    print(i, ':', df[i].nunique())
```

### Dropping Country, Postal Code

```python
df = df.drop(['Country', 'Postal Code'], axis = 1)
df.head()
```

### Profit

```python
round(df['Profit'].mean(), 2)
```

        The average value of Profit is $28.66, which means that the supermart is in profit.

### Ship Mode vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('Ship Mode')['Profit'].mean().plot.bar()
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Ship Mode', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Supermart is slightly getting more profit when the shipping mode is first class.

### Segment vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('Segment')['Profit'].mean().plot.bar()
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Segment', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Although each segment is producing profit for the supermart, it needs to improve the profit is consumer segment.

### Region vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('Region')['Profit'].mean().plot.bar()
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Region', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Supermart needs to give more attention to the Central Region. The profit is relatively less as compared to the
        other three regions.

### State vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('State')['Profit'].mean().sort_values(ascending = False).head(10).plot.bar()
plt.xticks(rotation = 90, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('State', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Above are the top 10 states with highest profit.

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('State')['Profit'].mean().sort_values().head(10).plot.bar()
plt.xticks(rotation = 90, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('State', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Above are the top 10 states with highest losses.

### City vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('City')['Profit'].mean().sort_values(ascending = False).head(50).plot.bar()
plt.xticks(rotation = 90, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('City', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

         Above are the top 50 cities with highest profit.

         Above are the top 100 cities with highest losses.

### Category vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('Category')['Profit'].mean().plot.bar()
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Category', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Profit is more with Technical category.
        Scope of improvement required in Furniture category.

### Sub-Category vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('Sub-Category')['Profit'].mean().sort_values(ascending = False).plot.bar()
plt.xticks(rotation = 90, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Sub-Category', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Profit is very high with Copiers while supermart is facing losses with supplies, bookcases and tables.

### Sales vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
sns.regplot(x = 'Sales', y ='Profit', data = df, line_kws = {'color': 'red'})
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Sales', fontsize = 15)
plt.ylabel('Profit', fontsize = 15)
plt.show()
```

         Profit is increasing with increase in sales.

### Quantity vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('Quantity')['Profit'].mean().plot.bar()
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Quantity', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Profit is generally increasing with increase in quantity. Supermart needs to up-sale.

### Discount vs Profit/Loss

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
df.groupby('Discount')['Profit'].mean().plot.bar()
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Discount', fontsize = 15)
plt.ylabel('Average Profit', fontsize = 15)
plt.show()
```

        Profit is more when the discount is generally between 0-20%.
